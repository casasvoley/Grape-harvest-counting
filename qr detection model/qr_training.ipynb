```python
from ultralytics import YOLO
import torch
torch.cuda.is_available()
```

```python
# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model 
```

```python
# Train the model
results = model.train(data="qr.yaml", imgsz=800, augment=True, dropout=0.3, patience=25, momentum=0.9, weight_decay=0.0001, lr0=0.0001, lrf=0.00001, batch=2, epochs=100, plots=True, device=0, val=True, optimizer='Adam')
```
